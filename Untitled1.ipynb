```python
%matplotlib inline
from matplotlib import style
style.use("fivethirtyeight")
import matplotlib.pyplot as plt
    
```

```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
from sqlalchemy import func
```

```python
#Reflect Tables into SQLAlchemy ORM
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import create_engine, inspect
```

```python
#create engine and import sqlite file
engine=create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base=automap_base()
# reflect the tables
Base.prepare(engine,reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#Exploratory Climate Analysis
```

```python
## Design a query to retrieve the last 12 months of precipitation data and plot the results
#finding columns names
first_row=session.query(Measurement).first()
first_row.__dict__
```

```python
# Calculate the date 1 year ago from the last data point in the database
#calculating the most recent date
most_recent_date=session.query(Measurement.date)\
                    .order_by(Measurement.date.desc()).first()
most_recent_date
# Calculate the date 1 year ago from the last data point in the database
year_ago_date=dt.date(2017,8,23)-dt.timedelta(days=365)
print(year_ago_date)
```

```python
# Perform a query to retrieve the data and precipitation scores
data=session.query(Measurement.date, Measurement.prcp , Measurement.station, Measurement.tobs,Measurement.id).\
    filter(Measurement.date > '2016-08-23').\
    order_by(Measurement.date).all()
data
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
inspector=inspect(engine)
columns=inspector.get_columns('Measurement')
for c in columns:
    print(c["name"],c["type"])
```

```python
# Sort the dataframe by date
results=session.query(Measurement.date,Measurement.prcp).\
    order_by(Measurement.date.desc()).all()
results
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
df = pd.DataFrame(results[:], columns=['date','prcp'])
df.set_index('date', inplace=True, )
df.head(10)
```

```python
#dropping date with NAN 

df.dropna(inplace=True)
df.head(20)
```

```python
df.iloc[::-1].plot.bar(title="Precipitation Vs date")
plt.title("Hawaii Precipitation Measurements", fontweight='bold',size=10)
plt.xlabel("Date", size=8)
plt.ylabel("Precipitation (inches)",  size=9)
plt.tight_layout()
plt.show()
plt.savefig('precipitation_amounts.png')
plt.show()
```

```python
 # Use Pandas to calcualte the summary statistics for the precipitation data
df.describe()
```

```python
#STATION ANALYSIS
inspector=inspect(engine)
columns=inspector.get_columns('station')
for c in columns:
    print(c["name"],c["type"])
```

```python
Total_Stations_Number=session.query(Measurement).group_by(Measurement.station).count()
print(Total_Stations_Number)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
session.query(Measurement.station,func.count(Measurement.station))\
.group_by(Measurement.station)\
.order_by(func.count(Measurement.station).desc()).all()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
most_active_station_details=session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == "USC00519281").all()
print(most_active_station_details)
```

```python
# Choose the station with the highest number of temperature observations.

results = session.query(Measurement.tobs).filter(Measurement.station == "USC00519281").filter(Measurement.date >= year_ago_date).all()
results
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
df = pd.DataFrame(results, columns=["tobs"])
df.plot.hist(bins=12)

plt.xlabel('Temperature Bins (12)', fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title(f"Temperature Results for station ",fontsize=12)
plt.tight_layout()
plt.savefig('temperature_results_hist.png')
plt.show()
```
